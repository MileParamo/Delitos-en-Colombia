# 3. Revisión de valores unicos

Inicialmente revisaremos los valores únicos de cada variable

```python
# Revisión de valores únicos
print("\n### Revisión de valores únicos")
for column in df_total.columns:
    unique_values = df_total[column].unique()
    print(f"Valores únicos en la columna '{column}':")
    print(unique_values)
    print("\n")
```

Del resultado se observan varios valores unicos repetidos, posiblemente por errores de tipificación o error humano en escritura, por lo que vamos a usar una libreería que identifica palabrás similares calculando un puntaje por caracter, adicionalmente colocaremos un treshhold de 90 para posteriormente unificarlas en la categoria con mas cantidad de matches

```python
from fuzzywuzzy import process

def unify_similar_values(column, threshold=90):
    unique_values = df_total[column].unique()
    for value in unique_values:
        matches = process.extract(value, unique_values, limit=None)
        for match in matches:
            if match[1] >= threshold and match[0] != value:
                df_total[column] = df_total[column].replace(match[0], value)

# Unificar valores similares en las columnas 'Departamento', 'Municipio' y 'Genero'
unify_similar_values('Departamento')
unify_similar_values('Municipio')
unify_similar_values('Genero')

# Revisión de valores únicos
print("\n### Revisión de valores únicos")
for column in df_total.columns:
    unique_values = df_total[column].unique()
    print(f"Valores únicos en la columna '{column}':")
    print(unique_values)
    print("\n")
```

Si bien nos ayudo a corregir una gran cantidad de valores aún quedan unos pocos que corregiremos manualmente

```python
# Unificar valores en la columna 'Departamento'
df_total['Departamento'] = df_total['Departamento'].replace(['-', 'No Registra'], 'No Reportado')

# Unificar valores en la columna 'Municipio'
df_total['Municipio'] = df_total['Municipio'].replace(['no registrado', '-'], 'No Reportado')

# Unificar valores en la columna 'Genero'
df_total['Genero'] = df_total['Genero'].replace(['-', 'No Registra'], 'No Reportado')


# Revisión de valores únicos
print("\n### Revisión de valores únicos")
for column in df_total.columns:
    unique_values = df_total[column].unique()
    print(f"Valores únicos en la columna '{column}':")
    print(unique_values)
    print("\n")
```

# 8. Detección de Outliers

Realizaremos un análisis básico de detección de otliers

```python
# Usar el método IQR para detectar outliers en la columna 'Cantidad Del Delito'
Q1 = df['Cantidad Del Delito'].quantile(0.25)
Q3 = df['Cantidad Del Delito'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Cantidad Del Delito'] < (Q1 - 1.5 * IQR)) | (df['Cantidad Del Delito'] > (Q3 + 1.5 * IQR))]
print("\n### Detección de outliers en 'Cantidad Del Delito'")
outliers
```

Para este caso y sabiendo la naturaleza de los datos podemos definir que si bien el metodo de IQR nos pude dar outliers basado en 3 desviaciones estandar, no es aplicable de gran manera a esta información ya que una gran cantidad de datos están recostados sobre el valor 1 que es el mas frecuente, por lo que todo lo que sea diferente seguramente no alcance a estar dentro de las 3 desviaciones, omitiremos cualqueir tipo de proceso sobre estos "Outliers" dando así fin al análisis de calidad de los datos
